```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load the data
df = pd.read_csv('matric_results.csv')
```

```python
df.head(5)
```

```python
# check the structure of the data
print(df.head())
print(df.tail())
print(df.info())
```

```python
# check for missing data
print(df.isnull().sum())

# remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# descriptive statistics
print(df.describe())
```

```python
# univariate analysis
df['quintile'].hist()
plt.title('Distribution of Quintile')
plt.xlabel('Quintile')
plt.ylabel('Frequency')
plt.show()

sns.boxplot(x='quintile', y='perc_2016', data=df)
plt.title('Distribution of Percentage Pass by Quintile')
plt.xlabel('Quintile')
plt.ylabel('Percentage Pass')
plt.show()
```

```python
# bivariate analysis
sns.scatterplot(x='wrote_2016', y='passed_2016', data=df)
plt.title('Relationship between Number of Learners who Wrote and Passed in 2016')
plt.xlabel('Number of Learners Who Wrote')
plt.ylabel('Number of Learners Who Passed')
plt.show()

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# multivariate analysis
sns.pairplot(df, hue='quintile')
plt.show()
```
